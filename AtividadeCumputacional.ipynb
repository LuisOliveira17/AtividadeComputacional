# *Achando a solução ótima para a Função Objetivo MaxL=100x1+80x2:*


Este notebook apresenta a resolução do estudo de caso “Otimização de Alocação de Desenvolvedores”, proposto na disciplina de Cálculo.
O objetivo é determinar a melhor distribuição de horas entre dois projetos da empresa fictícia TechSolutions — o Projeto Alpha e o Projeto Beta — de modo a maximizar o lucro total, respeitando as restrições de capacidade técnica e regras de gerenciamento.

Primeiramente, o cenário conta com restrições e requisitos mínimos, gerando assim, inequações de formação:

image.png

Cada uma dessas inequações foi representada graficamente (utilizando o *GeoGebra* ) no plano cartesiano, delimitando regiões de possível atuação. A interseção entre todas essas restrições forma a chamada região factível, que representa o conjunto de todas as combinações de horas dedicadas aos projetos Alpha e Beta que satisfazem simultaneamente as limitações impostas:

image.png



Com base nos pontos de formação, basta aplica-los na Função Objetivo princípal para assim calcular o lucro obtido a partir de cada ponto:

```python
#Definindo os pontos que formam a área Factível:
listaPontos={
    "a":[0,266.67],
    "b":[80,240],
    "d":[120,200],
    "c":[160,100]
}
```

```python
#Definindo Função Objetivo:
def Func(x1,x2):
  return 100*x1+80*x2

#x1 representa o total de horas dedicadas ao projeto Alpha e x2, ao projeto Beta
```

```python
# Aplicando os pontos da Função e gerando os cenários:
#Só será testado os pontos que atendem as restrições

resultados = {}
for nome, (x1, x2) in listaPontos.items():
    if x1 >= 40 and x2 >= 145 and (x1 + x2) <= 350:
        resultados[nome] = Func(x1, x2)
```

```python
#Mostrando os resultados e dizendo qual a solução ótima:

for ponto, valor in resultados.items():
    print(f"Ponto {ponto}:{valor:.2f}")

pontoOtimo= max(resultados, key=resultados.get)
x1, x2 = listaPontos[pontoOtimo]
lucro = resultados[pontoOtimo]

print("\nPortanto, o ponto ótimo que visa maximizar o lucro é:",pontoOtimo)
print(f"Sendo eles respectivamente:{x1},{x2}")
print(f"Gerando um lucro bruto de:{lucro} R$")
```

# *Analisando os gargalos*
Como sabemos os pontos da solução ótima (120,200), basta substituirmos nas funções de cada setor da empresa para analisarmos se há sobra ou não:


```python
x1 = 120
x2 = 200

# Codificação
resultadoCod = 0.4 * x1 + 0.2 * x2
limiteCod = 120
print("Codificação:", resultadoCod, "<=", limiteCod)
if resultadoCod == limiteCod:
    print(" Está no limite (Gargalo)!\n")
else:
    print(" Ainda há folga nessa restrição.\n")

# Análise e Design
resultadoAna = 0.1 * x1 + 0.3 * x2
limiteAna = 80
print("Análise e Design:", resultadoAna, "<=", limiteAna)
if resultadoAna == limiteAna:
    print("Está no limite (Gargalo)!\n")
else:
    print("Ainda há folga nessa restrição.\n")

# Testes
resultadoTestes = 0.5 * x1 + 0.2 * x2
limiteTestes = 100
print("Testes:", resultadoTestes, "<=", limiteTestes)
if resultadoTestes == limiteTestes:
    print(" Está no limite (Gargalo)!\n")
else:
    print("Ainda há folga nessa restrição.\n")

# Total de horas
resultadoTotal = x1 + x2
limiteTotal = 350
print("Total de horas:", resultadoTotal, "<=", limiteTotal)
if resultadoTotal == limiteTotal:
    print("Está no limite (Gargalo)!\n")
else:
    print("Ainda há folga nessa restrição.\n")

# Mínimo Alpha
resultadoMinAlpha = x1
limiteMinAlpha = 40
print("Mínimo Projeto Alpha:", resultadoMinAlpha, ">=", limiteMinAlpha)
if resultadoMinAlpha == limiteMinAlpha:
    print("Está exatamente no mínimo permitido!\n")
else:
    print("Está acima do mínimo exigido.\n")

# Mínimo Beta
resultadoMinBeta = x2
limiteMinBeta = 100
print("Mínimo Projeto Beta:", resultadoMinBeta, ">=", limiteMinBeta)
if resultadoMinBeta == limiteMinBeta:
    print("Está exatamente no mínimo permitido!\n")
else:
    print("Está acima do mínimo exigido.\n")

```

# *CONCLUSÃO FINAL:*

Com os valores ótimos de x₁ = 120 horas destinadas ao Projeto Alpha e x₂ = 200 horas destinadas ao Projeto Beta, a empresa obtém um lucro máximo de R$ 28.000,00.

A análise das restrições mostra que apenas a restrição de Testes atingiu o seu limite, caracterizando-se como o principal gargalo do sistema. Já as restrições de Codificação, Análise e Design e Total de Horas (<= 350) apresentaram folga, o que indica que esses recursos não foram utilizados em sua capacidade máxima.

Além disso, as restrições mínimas exigidas para os projetos Alpha e Beta foram respeitadas, garantindo o equilíbrio entre os dois trabalhos.

Portanto, conclui-se que a atividade de Testes é o recurso crítico que limita o aumento do lucro total. Para ampliar o desempenho da empresa, recomenda-se aumentar a capacidade de Testes ou realocar parte dos recursos ociosos (como codificação e design) de forma estratégica, otimizando a distribuição das horas de trabalho entre as equipes.
